```python
import requests
import json
import pandas as pd
import numpy as np
import time
from io import StringIO
from bs4 import BeautifulSoup
import re
```

```python
kecamatan = pd.read_csv("./kecamatan.csv")
kecamatan = kecamatan[kecamatan['kode_kab'] == 1604]
kecamatan.head()
```

```python
kecamatan.shape
```

```python
for k in kecamatan['kecamatan_display']:
    url = 'https://www.google.com/maps/search/{}'.format(k.replace(" ", "%20"))
    r = requests.get(url)

    content = BeautifulSoup(r.content)
    text_to_find = content.findAll(string=re.compile("@-"))
    latlong = re.search("@([^,]*,[^,]*),", string=text_to_find[0])
    print(latlong.group(1))
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="chromebook")
```

```python
print(geolocator.geocode("Mulak Sebingkai, Lahat, Sumatera Selatan"))
```

```python
for i, k in kecamatan.iterrows():
    print(k['kecamatan_display'])
```

```python
kecamatan['lat'] = None
kecamatan['long'] = None
```

```python
for i, k in kecamatan.iterrows():
    location = geolocator.geocode(k['kecamatan_display'].replace("Kecamatan ", "") + ", Lahat, Sumatera Selatan")
    # print(location.latitude, location.longitude)
    lat = None
    long = None
    if location is not None:
        lat = location.latitude
        long = location.longitude
    if k['kecamatan_display'] == "Kecamatan Mulak Sebingkai":
        lat = -3.9783251
        long = 103.5206075
    kecamatan.at[i, 'lat'] = lat
    kecamatan.at[i, 'long'] = long
    print("{}: {}, {}".format(k['kecamatan_display'], lat, long))
```
