```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
kecamatan = pd.read_csv("/content/kecamatan.csv")
kecamatan = kecamatan[kecamatan['kode_kab'] == 1604]
kecamatan.head()
```

```python
kecamatan.shape
```

```python
kecamatan['lat'] = None
kecamatan['long'] = None
kecamatan.head()
```

```python
for i, k in kecamatan.iterrows():
    url = 'https://www.google.com/maps/search/{}'.format(k['kecamatan_value'].replace(" ", "%20"))
    r = requests.get(url)

    content = BeautifulSoup(r.content)
    text_to_find = content.findAll(string=re.compile("@-"))

    lat = None
    long = None

    if len(text_to_find) > 0:
        latlong = re.search("@([^,]*,[^,]*),", string=text_to_find[0]).group(1).split(",")
        lat = latlong[0]
        long = latlong[1]
    if k['kecamatan_display'] == "Kecamatan Mulak Sebingkai":
        lat = -3.9783251
        long = 103.5206075
    kecamatan.at[i, 'lat'] = lat
    kecamatan.at[i, 'long'] = long

    print("{}: {}, {}".format(k['kecamatan_display'], lat, long))
    # print(latlong.group(1))
    # break
```
